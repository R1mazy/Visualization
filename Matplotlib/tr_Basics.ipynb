```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

```python
x
```

```python
y
```

```python
plt.plot(x, y) 
plt.xlabel('Название оси X')
plt.ylabel('Название оси Y')
plt.title('Название графика')
plt.show() # Требуется если не в jupyter; 
```

```python
plt.plot(x, y) 
plt.xlabel('Название оси X')
plt.ylabel('Название оси Y')
plt.title('Название графика')
plt.xlim(0,6)
plt.ylim(0,12)
plt.show()
```
